```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("power_consumption_in_india/long_data_.csv")
df.Dates = pd.to_datetime(df.Dates, dayfirst = True)
```

```python
df.set_index('Dates', inplace = True)
```

```python
df = df.pivot_table(values = 'Usage', columns = 'States', index = df.index)
```
