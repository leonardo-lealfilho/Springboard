```python
import pandas as pd
```

```python
df = pd.read_csv("Indicators.csv", index_col = ['Year', 'Stock'])
```

```python
df.head()
```

```python
# Getting the dummies variables for the categorical data under the Sector feature
df = pd.get_dummies(df)
```

```python
df.head()
```

```python
df.describe()
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Isolating the Features
X = df.drop(['Class'], axis = 1)

# Isolating the Label
y = df.Class
```

```python
# Separating the training and the testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Instatiating StandardScaler
SS_scaler = StandardScaler()
```

Only features 0 to 19 can be standardized because the rest of the features are meant to be boolean values intended to describe whether the stock contains that feature or not.

```python
# Fitting to the training data to be scaled
SS_scaler.fit(X_train.iloc[:,0:19])
```

```python
# Separating the names for the new columns with the scaled data
scaled_col_names = []
for col in X.iloc[:, 0:19].columns:
    scaled_col_names.append("Scaled " + col)
```

```python
# Adding the scaled columns to the training dataset under their new column names
X_train[scaled_col_names] = SS_scaler.transform(X_train.iloc[:,0:19])
```

```python
# Checking the new training dataset
X_train.describe()
```

```python
# Fitting to the testing data to be scaled
SS_scaler.fit(X_test.iloc[:,0:19])
```

```python
# Adding the scaled columns to the testing dataset under their new column names
X_test[scaled_col_names] = SS_scaler.transform(X_test.iloc[:,0:19])
```
